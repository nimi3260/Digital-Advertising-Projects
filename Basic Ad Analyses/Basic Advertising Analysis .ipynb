```python
#importing the packages necessary to run the analyses 
import pandas
import pip
import arrow
import numpy as np
import statsmodels.api as sm 
```

```python
#this assigns the filename we're trying to load in to a string variable
trav = 'Travel Pony Facebook.csv'
```

```python
#this uses the pandas package to load the csv file into a pandas dataframe
df = pandas.read_csv(trav) 
```

```python
#creating a new column in the data frame
df['cost_per_impression'] = df['Amount Spent (USD)'] / df['Impressions'] 
```

```python
#In our dataframe, we have a column called 'Start Date'. 
#We want to extract the day of the week (ex: Moday, Tuesday, Wednesday, etc.)  
#First we initialized an empty list, and made a for loop to iterate through each row in the dataframe and extract the day of the week 
#We then assign the day of the week to a new column in our dataframe 
dayofweek= []

for row in range(len(df)):
    date = arrow.get(df['Start Date'][row], 'M/D/YY') 
    dayofweek.append(date.weekday()) 

df['Week Day'] = dayofweek
```

```python
#We want to find which day of the week works best/worst for our variable 'cost per impression'
#The code below is working the same as a pivoting data function to sort the data by the day of the week  
group = df.groupby('Week Day')  
group.aggregate(np.mean)
```

```python
#When considering 'cost per impression', what day of the week works best? (What day is it cheapest to generate impressions)  
#Saturday is the cheapest, thus it is the day that works best. 

#What day works worst? (What day is most expensive)
#Friday is the day that works the worst becasue it is the most expensive. 
```

```python
#Next we want to compute a correlations between 'Amount Spent' and other variables: 
#Reach 
print( df[['Amount Spent (USD)','Reach']].corr())

#Frequecy 
print( df[['Amount Spent (USD)','Frequency']].corr())

#Unique Clicks 
print( df[['Amount Spent (USD)','Unique Clicks']].corr())

#Page Likes
print( df[['Amount Spent (USD)','Page Likes']].corr())
```

```python
#Which correlation is the strongest? What does that mean practically? 

#There is a strong positive correlation between Amount Spent and Unique Clicks. 
#As the number of USD Spent increases, so does the number of Unique Clicks. 
```

```python
#performing a simple multiple regression analysis where Unique Clicks is the dependent variable and Reach and Frequency are the independent (predictor) variables.
X = df[['Reach','Frequency']]
Y = df['Unique Clicks']

model = sm.OLS(Y, X).fit()
predictions = model.predict(X) 

print (model.summary())
```

```python
#What variable most strongly predicts unique clicks? What does that mean practically? 

#Frequency is the variable that most strongly predicts Unique Clicks. 
#It is statistically significant, and the coefficient indicates that for every 
#unit increase in frequency you can expect unique clicks to increase by an 
#average of 3.6.
```
